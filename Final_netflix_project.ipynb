```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\DELL\Desktop\Projects\Netflix\netflix_titles.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df['country']= df['country'].fillna(df['country'].mode()[0])
```

```python
df['director'].replace(np.nan,'No Data',inplace=True)
df['cast'].replace(np.nan,'No Data',inplace=True)
df.dropna(inplace=True)

```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format="%B %d, %Y", errors='coerce')


df['Year_added']=  df['date_added'].dt.year
df['Month_added']= df['date_added'].dt.month
df['Month_added']= df['date_added'].dt.month_name()
```

```python
df
```

```python
df['Year_added'] = df['Year_added'].fillna(0).astype(int)
```

```python
df.head()
```

```python
countofcontent=df['type'].value_counts().reset_index()
countofcontent
```

```python
plt.pie(countofcontent['count'],labels=['Movie','TvShow'],autopct='%1.0f%%',explode=[0,0.1],shadow=True)
plt.title('Distribution of Movie and TvShow')
```

```python
contentaddedbyyear=df.groupby(['Year_added'])['type'].value_counts().sort_values(ascending=False).reset_index().head(12)
contentaddedbyyear
```

```python
sns.barplot(x='Year_added',y='count',hue='type',data=contentaddedbyyear)
plt.xticks(rotation=45)
plt.title('Content uploaded on Netflix by Year')
plt.xlabel('Year')
```
